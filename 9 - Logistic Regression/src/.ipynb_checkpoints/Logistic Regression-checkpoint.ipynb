```python
# -*- coding: utf-8 -*-
"""
Created on Thu Jul 16 21:15:55 2020

@author: Hossein
"""

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


class CommandCenter :
    testSetSize = 0.8
    trainSetSize = 1 - testSetSize
    A_Class_Train_X = []
    A_Class_Train_Y = []
    B_Class_Train_X = []
    B_Class_Train_Y = []
    Probability = 0.5

def readData(fileName):

    wholeData = pd.read_csv(fileName)
    wholeData = wholeData.drop(columns=['Z', 'ZZ'])
    wholeData = wholeData[wholeData.Result != "Iris-versicolor"]
    wholeData["Result"] = wholeData["Result"].apply(lambda x : 1 if x == "Iris-setosa" else 0 )
    return wholeData

def sigmoid(z):
    return 1 / (1 + np.exp(-z))


def h(features, thetas):
    z = np.dot(thetas.T, features )
    print(z)
    H = sigmoid(z)
    #print(H)
    return H

def costFunction(X,thetas , y):
    summation = (y * np.log(h(X,thetas)) ) + ((1 - y) * (np.log( 1 - h(X,thetas))))  
    return summation 

def updateThetas ( thetas , m , X , Y):
    #print("Thetas Shape : "+ str(thetas.shape))
    Bravo = h(X,thetas) 
    Bravo = Bravo.reshape(1,1)
    #print("Shape Bravo : " + str(Bravo.shape))
    Bravo -= Y
    #print(Y)
    ###
    if ( Y == 1 ):
        CommandCenter.A_Class_Train_X.append(Bravo)
        CommandCenter.A_Class_Train_Y.append(Y)
    else:
        CommandCenter.B_Class_Train_X.append(Bravo)
        CommandCenter.B_Class_Train_Y.append(Y)
    ###
    
    X = X.reshape(2,1)
    #print("Shape X : " + str(X.reshape(2,1).shape))
    Charlie = ( X.T * Bravo) /m
    Charlie = Charlie.reshape(2,1)
    #print('==========')
    thetas = thetas - Charlie
    thetas = thetas.reshape(2,1)
    return thetas

def split( primary_perc , secondary_perc , dataSet ):
    # 0 <=  primary <= 1 
    # 0 <= secondary <= 1
    # primary + secondary = 1
    
    dataSet = dataSet.sample(frac=1) # shuffle Data
    
    primaryLen = len(dataSet) * primary_perc
    primary = dataSet.iloc[0:int(primaryLen)]

    #secondaryLen = len(dataSet) * secondary_perc
    secondary = dataSet.iloc[ int(primaryLen) : ]
    return primary , secondary
      
def initialize(DataSet):
    #initialize theta vector
    numberOfFeatures = len(DataSet.iloc[0]) - 1 
    Thetas = np.zeros((numberOfFeatures,1))
    return Thetas


    
    
def train(trainSet):

    trainSetLength = len(trainSet)

    Thetas = initialize(trainSet)
    
    
    m = trainSetLength
    
    # Create X (all the feature columns)
    X = np.array(trainSet.drop("Result", axis=1))

    # Create y (the target column)
    Y = np.array(trainSet["Result"])

    
    
    for i in range(trainSetLength):
        index = len(X[i])
        X_sample = np.array(X[i])
        Y_sample = np.array(Y[i])
        X_sample.reshape(index,1)
        Y_sample.reshape(1,1)
        Thetas = updateThetas(Thetas,m,X_sample,Y_sample)
  
    return Thetas 

def main():
    dataSet = readData("iris.csv")
    trainSet , testSet = split(CommandCenter.trainSetSize , CommandCenter.testSetSize , dataSet)
    thetas = train(trainSet)
    #print(" Thetas : " + str(thetas))
    xxx = []
    yyy = []

        
    
    #for item in CommandCenter.A_Class_Train_X :
        #xxx.append(float(item))
    #for item in CommandCenter.A_Class_Train_Y :
        #yyy.append(float(item))
    plt.scatter(xxx , yyy,color='black' , label = "A")
    #plt.plot(1,2)
    plt.xlabel('x - axis') 
    # naming the y axis 
    plt.ylabel('y - axis') 
    # giving a title to my graph 
    plt.title('ccc') 
    # show a legend on the plot 
    plt.show()
main()
```
